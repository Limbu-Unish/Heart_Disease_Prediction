# Data Cleaning

```python
# Creating the copy of the orginal df for proper backup
df = df_initial.copy()
```

```python
# Correcting the format of the column names
df.columns = df.columns.str.replace(' ', '_').str.strip()
```

```python
# Cleaning the Age variable
df['Age'] = df['Age'].fillna(df['Age'].median()).astype(int)
```

Successfully cleaned the Age Variable

```python
# Cleaning the Blood Pressure variable
df['Blood_Pressure'] = df['Blood_Pressure'].fillna(df['Blood_Pressure'].median()).astype(int)
```

Successfully cleaned the Blood Pressure variable

```python
# Cleaning the Cholesterol Level variable
df['Cholesterol_Level'] = df['Cholesterol_Level'].fillna(df['Cholesterol_Level'].median()).astype(int)
```

Successfully cleaned the Cholesterol Level variable

```python
# Cleaning the Gender variable
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

Successfully cleaned the Gender variable

```python
# Cleaning the Excersie Habit variable
df['Exercise_Habits'] = df['Exercise_Habits'].fillna(df['Exercise_Habits'].mode()[0])
```

Successfully cleaned the Excersie habit variable

```python
# Cleaning the smoking variable
df['Smoking'] = df['Smoking'].fillna(df['Smoking'].mode()[0])
```

Successfully cleaned the Smoking variable

```python
# Cleaning the Family heart disease variable
df['Family_Heart_Disease'] = df['Family_Heart_Disease'].fillna(df['Family_Heart_Disease'].mode()[0])
```

Successfully cleaned the Family heart disease variable

```python
# Cleaning the Diabetes variable
df['Diabetes'] = df['Diabetes'].fillna(df['Diabetes'].mode()[0])
```

Successfully cleaned the Diabetes variable

```python
# Cleaning the BMI varaible
df['BMI'] = df['BMI'].fillna(df['BMI'].mode()[0]).astype(float)
```

```python
df['BMI'] = round(df['BMI'], 1)
```

Successfully cleaned the BMI variable

```python
# Cleaning the remaining variables
categorical_variables = {
    'High_Blood_Pressure',
    'Low_HDL_Cholesterol',
    'High_LDL_Cholesterol',
    'Alcohol_Consumption',
    'Stress_Level',
    'Sugar_Consumption'
}

continuos_variables = {
    'Sleep_Hours',
    'Triglyceride_Level',
    'Fasting_Blood_Sugar',
    'CRP_Level',
    'Homocysteine_Level',
}

# Cleaning the categorical variables
for var in categorical_variables:
    if var in df.columns:
        df[var] = df[var].fillna(df[var].mode()[0])
    else:
        print('The variables were not found.')

# Cleaning the continuos variables
for var in continuos_variables:
    if var in df.columns:
        df[var] = df[var].fillna(df[var].median())
    else:
        print('The variables were not found.')
```

```python
df['Sleep_Hours'] = df['Sleep_Hours'].astype(float)
df['Triglyceride_Level'] = df['Triglyceride_Level'].astype(int)
df['Fasting_Blood_Sugar'] = df['Fasting_Blood_Sugar'].astype(int)
df['CRP_Level'] = round(df['CRP_Level'], 1)
df['Homocysteine_Level'] = round(df['Homocysteine_Level'], 1)
```

Successfully cleaned the remaining variables

```python
# Performing encoding in the categorical variables for performing further process
# For Gender variable
df['Gender'] = df['Gender'].map({'Male': 0, 'Female':1})
```

```python
# For Smoking, Family_Heart_Disease, Diabetes, High_Blood_Pressure, Low_HDL_Cholesterol, High_LDL_Cholesterol
Yes_No_Columns = [
    'Smoking',
    'Family_Heart_Disease',
    'Diabetes',
    'High_Blood_Pressure',
    'Low_HDL_Cholesterol',
    'High_LDL_Cholesterol'
]

map_values = {'Yes': 1, 'No': 0}
for col in Yes_No_Columns:
    df[col] = df[col].map(map_values)
```

```python
# For Exercise_Habits, Alcohol_Consumption, Stress_Level, Sugar_Consumption
from sklearn.preprocessing import OrdinalEncoder

H_M_L_Columns = [
    'Exercise_Habits',
    'Alcohol_Consumption',
    'Stress_Level',
    'Sugar_Consumption'
]

H_M_L_Order = ['Low', 'Medium', 'High']
H_M_L_Categories = [H_M_L_Order for _ in H_M_L_Columns]
H_M_L_Encoder = OrdinalEncoder(categories=H_M_L_Categories)
H_M_L_Encoded_Array = H_M_L_Encoder.fit_transform(df[H_M_L_Columns])
df[H_M_L_Columns] = pd.DataFrame(H_M_L_Encoded_Array, columns=H_M_L_Columns, index=df.index).astype(int)
```

```python
# For the target variable
from sklearn.preprocessing import LabelEncoder

target_encoder = LabelEncoder()
df['Heart_Disease_Status'] = target_encoder.fit_transform(df['Heart_Disease_Status'])
```

```python
df['Sleep_Hours'] = round(df['Sleep_Hours'],1)
```
